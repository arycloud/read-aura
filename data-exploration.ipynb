```python
from statistics import correlation

import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(f"{path}/books.csv")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = plt.figure(figsize=(12, 6))
axis = plt.gca()
# axis = plt.axis()
# sns.heatmap(df.isna().transpose(), cbar=False, ax=axis)
# create heatmap of missing values
sns.heatmap(df.isna().transpose(), ax=axis)
plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

# Findings

1. We found that the above columns have less than 5% missing values, so we can remove them
2. Below we did that by using the negation logic and save the dataset into 'book_missing' variable

```python
import numpy as np
```

```python
df['missing_descriptions'] = np.where(df['description'].isna(), 1, 0)
df['age_of_book'] = 2025 - df['published_year']
```

```python
columns_of_interest = ['num_pages', 'age_of_book', 'missing_descriptions','average_rating']
correlation_matrix =  df[columns_of_interest].corr(method = 'spearman')
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": correlation_matrix})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

## Findings

1. Found that the above columns have less then 5% of missing values, so we can remove those data samples ...
2. Below we did that, just remove the data instances where we have missing values for these colums using a negation logic...

```python
book_missing = df[ ~(df['description'].isna()) &
                   ~(df['num_pages'].isna()) &
                   ~(df['average_rating'].isna()) &
                   ~(df['published_year'].isna())
]

book_missing
```

```python
book_missing['categories'].value_counts().reset_index().sort_values("count", ascending=False)
```

### Above we can notice that skewed distribution of books categories, most of them belongs to fiction, so we will fix that using LLM

```python
book_missing
```

```python
book_missing['words_in_description'] = book_missing['description'].str.split().str.len()
book_missing
```

```python
book_missing.loc[book_missing['words_in_description'].between(15, 24), "description"]
```

```python
book_missing.loc[book_missing['words_in_description'].between(25, 34), "description"]
```

```python
book_missing_25_words = book_missing[book_missing['words_in_description'] >= 25]
book_missing_25_words
```

```python
book_missing_25_words['title_and_subtitle'] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)

book_missing_25_words['title_and_subtitle']
```

```python
book_missing_25_words = book_missing_25_words.copy()
book_missing_25_words["title_and_subtitle"] = book_missing_25_words.apply(
    lambda row: row["title"] if pd.isna(row["subtitle"]) else f"{row['title']}: {row['subtitle']}",
    axis=1
)

book_missing_25_words
```

```python
book_missing_25_words['tagged_description'] = book_missing_25_words[ ['isbn13', 'description']].astype(str).agg(" ".join, axis=1)
book_missing_25_words
```

```python
# now, lets save the cleaned dataset

(
    book_missing_25_words
    .drop(['subtitle', 'missing_descriptions', 'age_of_book', 'words_in_description'], axis=1)
    .to_csv("book_cleaned.csv", index=False)
)
```
